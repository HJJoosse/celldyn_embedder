## Graphing varying sample sizes with performance (as well as knn number)

```python
sample_n_n_5 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_tune_MF100_sizes_large_n_n_5_std.csv")
sample_n_n_15 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_tune_MF100_sizes_large_n_n_15_std.csv")
sample_n_n_30 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_tune_MF100_sizes_large_n_n_30_std.csv")
sample_n_n_50 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_tune_MF100_sizes_large_n_n_50_std.csv")
sample_n_n_100 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_tune_MF100_sizes_large_n_n_100_std.csv")

sample_ls = [sample_n_n_5,sample_n_n_15, sample_n_n_30, sample_n_n_50,sample_n_n_100]
```

```python
sample_n_n_50.columns
```

```python
def sizes_against_performance_plot(sample_ls,performance_column, y_title):
    sizes = sample_ls[0]["sample_size"].values
    for sample in sample_ls:
        sns.lineplot(data=sample,x=sample["sample_size"], y=sample[performance_column])
    
    plt.legend([5,15,30,50,100], title = "Nearest Neighbours", title_fontsize=8, loc= 'lower right')
    plt.xlabel('Sample size', fontsize=10)
    plt.ylabel(y_title, fontsize=10)
    plt.show()


```

```python
sizes_against_performance_plot(sample_ls,"Random_triplets", "Random triplet score")
```

```python
sizes_against_performance_plot(sample_ls,"duration_in_second", "Duration in seconds")
```

```python
sizes_against_performance_plot(sample_ls,"neighbor_kept_ratio", "Neighbourhood kept ratio score")
```

```python
sizes_against_performance_plot(sample_ls,"neighbor_kept_15", "Neighbourhood kept ratio (default setting) score")
```

```python
sizes_against_performance_plot(sample_ls,"Trustworthiness", "Trustworthiness score")
```

```python
sizes_against_performance_plot(sample_ls,"dcor_1", "Data distance correlation score")
```

```python
sizes_against_performance_plot(sample_ls,"dcor_2", "Distance correlation score")
```

## Graphing embedding results

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sample_5000 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_hyper_op_results_5000_MF100_n_n.csv")
sample_10000 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_hyper_op_results_10000_MF100_n_n.csv")
sample_20000 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_hyper_op_results_20000_MF100_n_n.csv")
sample_40000 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_hyper_op_results_40000_MF100_n_n.csv")
sample_60000 = pd.read_csv("models/embedding/UMAP/cell_dyn_UMAP_hyper_op_results_60000_MF100_n_n.csv")
sample_ls = [sample_5000,sample_10000, sample_20000, sample_40000,sample_60000]
```

```python
sample_n_n_5 = []
sample_n_n_15 = []
sample_n_n_30 = []
sample_n_n_50 = []
sample_n_n_100 = []
for ls in sample_ls:
    sample_n_n_5.extend(ls[ls['n_neighbors'] == 5]["Distance correlation"].values)
    sample_n_n_15.extend(ls[ls['n_neighbors'] == 15]["Distance correlation"].values)
    sample_n_n_30.extend(ls[ls['n_neighbors'] == 30]["Distance correlation"].values)
    sample_n_n_50.extend(ls[ls['n_neighbors'] == 50]["Distance correlation"].values)
    sample_n_n_100.extend(ls[ls['n_neighbors'] == 100]["Distance correlation"].values)
```

```python
"""
import matplotlib 
font = {'family' : 'normal',
        'weight' : 'light',
        'size'   : 15}

SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12


matplotlib.rc('xtick', labelsize=BIGGER_SIZE) 
matplotlib.rc('ytick', labelsize=BIGGER_SIZE) 
matplotlib.rc('font', **font)
"""
sizes = [5000,10000,20000,40000,60000]
plt.style.use("seaborn-paper")
sns.lineplot(x=sizes, y=sample_n_n_5)
sns.lineplot(x=sizes, y=sample_n_n_15)
sns.lineplot(x=sizes, y=sample_n_n_30)
sns.lineplot(x=sizes, y=sample_n_n_50)
sns.lineplot(x=sizes, y=sample_n_n_100)
plt.legend([5,15,30,50,100], title = "Nearest Neighbours", title_fontsize=8, loc= 'lower right')
#plt.title("UMAP embedding random triplets scores with different sample\n sizes and numbers of nearest neighbours", size = 15)
plt.xlabel('Sample size', fontsize=10)
plt.ylabel('Distance correlation score', fontsize=10)

#plt.xlim([4000,90000])
```
